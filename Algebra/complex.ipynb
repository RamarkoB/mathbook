```python
from manim import *
import numpy as np
```

```python
scale = 6
step = 3

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING ConstantCartesian

class ConstantCartesian(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 2,
                "stroke_opacity": 0.35
            })

        func = FunctionGraph(
            lambda x: x**2 + 1,
            color = BLACK
        )
        func.stroke_width = 10
        

        self.add(numberplane, func)
```

```python
%%manim -v WARNING ConstantCartesian

class ConstantCartesian(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 2,
                "stroke_opacity": 0.35
            })

        func = FunctionGraph(
            lambda x: x**2  - 6*x + 10,
            color = BLACK
        )
        func.stroke_width = 10
        

        self.add(numberplane, func)
```
