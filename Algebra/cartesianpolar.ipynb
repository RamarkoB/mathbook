```python
from manim import *
import numpy as np
```

```python
scale = 6
step = 3

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING ConstantCartesian

class ConstantCartesian(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 2,
                "stroke_opacity": 0.35
            })

        func = FunctionGraph(
            lambda x: 3,
            color = BLACK
        )
        func.stroke_width = 10
        

        self.add(numberplane, func)
```

```python
%%manim -v WARNING LinearCartesian

class LinearCartesian(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 2,
                "stroke_opacity": 0.35
        })

        func = FunctionGraph(
            lambda x: x,
            color = BLACK
        )
        func.stroke_width = 10
        

        self.add(numberplane, func)
```

```python
%%manim -v WARNING TrigCartesian

class TrigCartesian(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 2,
                "stroke_opacity": 0.35
        })

        func = FunctionGraph(
            lambda x: np.sin(x),
            color = BLACK
        )
        func.stroke_width = 10
        

        self.add(numberplane, func)
```

```python
scale = 2
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
def Constant(theta):
    r = 1
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])

def Linear(theta):
    r = theta
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])

def Trig(theta):
    r = np.sin(theta)
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])
```

```python
%%manim -v WARNING ConstantPolar

class ConstantPolar(Scene):
    def construct(self):
        polarplane_pi = PolarPlane(
            azimuth_units="PI radians",
            size = scale * 2,
            radius_step = 1,
            x_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            y_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 1,
                "stroke_opacity": 0.3
            }
        )

        curl = ParametricFunction(Constant, t_range = np.array([0, TAU]), fill_opacity=0, stroke_width=3).set_color(BLACK)

        self.add(polarplane_pi, curl)
```
