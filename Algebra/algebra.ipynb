```python
from manim import *
import numpy as np
```

```python
# scale = 10
scale = 6
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING Square

config.frame_width = scale * 2

class Square(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale + 4, scale + 4, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        square = FunctionGraph(
            lambda t: t**2,
            color = BLACK
        )
        square.stroke_width = 10

        self.add(numberplane, square)
```

```python
%%manim -v WARNING Cubed

config.frame_width = scale * 2

class Cubed(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        cubed = FunctionGraph(
            lambda t: (t - 4) * t * (t + 4) * (1/12),
            color = BLACK
        )
        cubed.stroke_width = 10

        self.add(numberplane, cubed)
```

```python
%%manim -v WARNING SineLine

config.frame_width = scale * 2

class SineLine(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        sine = FunctionGraph(
            lambda x: np.cos(x),
            color = BLACK
        )
        sine.stroke_width = 10

        dot = Dot((0, 1, 0), color=BLACK, radius = 0.18)

        self.add(numberplane, sine, dot)
```

```python
%%manim -v WARNING InverseSquare

config.frame_width = scale * 2

class InverseSquare(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        inversesquare = ImplicitFunction(
            lambda x, y: y**2 - x - 4,
            color = BLACK,
            x_range = [-scale - 5, scale + 5],
            y_range = [-scale - 5, scale + 5],
        )
        inversesquare.stroke_width = 10

        dot1 = Dot((0, np.sqrt(4), 0), color=BLACK, radius = 0.18)
        dot2 = Dot((0, -np.sqrt(4), 0), color=BLACK, radius = 0.18)

        self.add(numberplane, inversesquare, dot1, dot2)
```

```python
%%manim -v WARNING Folium

config.frame_width = 20

class Folium(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        self.frame_width = scale * 2
        numberplane = NumberPlane(
            x_range = (-scale * 2, scale * 2, step),
            y_range = (-scale * 2, scale * 2, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale * (10 / scale)
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        folium = ImplicitFunction(
            lambda x, y: (x + 1) ** 3 + y ** 3 - 18 * (x + 1) * y,
            color = BLACK,
            x_range = [-scale - 5, scale + 5],
            y_range = [-scale - 5, scale + 5],
        )
        folium.stroke_width = 15

        dot1 = Dot((0, 4.25, 0), color=BLACK, radius = 0.27)
        dot2 = Dot((0, 0.056, 0), color=BLACK, radius = 0.27)
        dot3 = Dot((0, -4.25, 0), color=BLACK, radius = 0.27)

        self.add(numberplane, folium, dot1, dot2, dot3)
```
