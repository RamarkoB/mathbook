```python
from manim import *
import numpy as np
```

```python
scale = 6
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING Cubed

config.frame_width = scale * 2

class Cubed(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        cubed = FunctionGraph(
            lambda t: t**3,
            color = BLACK
        )
        cubed.stroke_width = 10

        self.add(numberplane, cubed)
```

```python
%%manim -v WARNING CubedRoot

config.frame_width = scale * 2

class CubedRoot(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        cubed = FunctionGraph(
            lambda t: np.cbrt(t),
            color = BLACK
        )
        cubed.stroke_width = 10

        self.add(numberplane, cubed)
```

```python
%%manim -v WARNING SquaredImplicit

config.frame_width = scale * 2

class SquaredImplicit(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        squared = ImplicitFunction(
            lambda x,y: y**2 - x,
            color = BLACK
        )
        squared.stroke_width = 10

        self.add(numberplane, squared)
```

```python
%%manim -v WARNING SquaredRoot

config.frame_width = scale * 2

class SquaredRoot(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        squared = FunctionGraph(
            lambda t: -np.sqrt(t),
            x_range=[0, 6],
            color = BLACK
        )
        squared.stroke_width = 10

        self.add(numberplane, squared)
```
