```python
from manim import *
import numpy as np
```

```python
scale = 6
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING VertStretch

config.frame_width = 20

class VertStretch(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-10, 10, step),
            y_range = (-10, 10, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        cubed = FunctionGraph(
            lambda t: (t - 4) * t * (t + 4) * (1/12),
            color = BLACK
        )
        cubed.stroke_width = 10

        strech = FunctionGraph(
            lambda t: 2 * (t - 4) * t * (t + 4) * (1/12),
            x_range=[-6, 6],
            color = BLACK
        )
        strech.stroke_width = 8
        dashed = DashedVMobject(strech, num_dashes=35, equal_lengths =True)

        

        self.add(numberplane, cubed, dashed)
```

```python
%%manim -v WARNING VertShift

config.frame_width = 20

class VertShift(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-10, 10, step),
            y_range = (-10, 10, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        cubed = FunctionGraph(
            lambda t: (t - 4) * t * (t + 4) * (1/12),
            color = BLACK
        )
        cubed.stroke_width = 10
        
        strech = FunctionGraph(
            lambda t: 2.5 + (t - 4) * t * (t + 4) * (1/12),
            x_range=[-8, 6],
            color = BLACK
        )
        strech.stroke_width = 8
        dashed = DashedVMobject(strech, num_dashes=35, equal_lengths =True)

        self.add(numberplane, cubed, dashed)
```

```python
%%manim -v WARNING HoriStretch

config.frame_width = 20

class HoriStretch(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-10, 10, step),
            y_range = (-10, 10, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        cubed = FunctionGraph(
            lambda t: (t - 4) * t * (t + 4) * (1/12),
            color = BLACK
        )
        cubed.stroke_width = 10

        strech = FunctionGraph(
            lambda t: ((.5 * t) - 4) *  .5 * t * ((.5 * t) + 4) * (1/12),
            x_range=[-10, 10],
            color = BLACK
        )
        strech.stroke_width = 8
        dashed = DashedVMobject(strech, dashed_ratio=(2/3))

        

        self.add(numberplane, cubed, dashed)
```
