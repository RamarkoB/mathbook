```python
from manim import *
import numpy as np
```

```python
scale = 3
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING Step

config.frame_width = scale * 2

class Step(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })
        
        
        steps = VGroup()
        filled_dots = VGroup()
        hollow_dots = VGroup()

        steps.add(Line(
        start=numberplane.c2p(-2, -2),
        end=numberplane.c2p(0, -2),
        color=BLACK,
        stroke_width=10
        ))
        filled_dots.add(Dot(numberplane.c2p(-2, -2), color=BLACK, radius=0.12))
        hollow_dots.add(Dot(numberplane.c2p(0, -2), color=BLACK, radius=0.12))
        hollow_dots.add(Dot(numberplane.c2p(0, -2), color=WHITE, radius=0.06))
        
        steps.add(Line(
        start=numberplane.c2p(0, 2),
        end=numberplane.c2p(2, 2),
        color=BLACK,
        stroke_width=10
        ))
        
        filled_dots.add(Dot(numberplane.c2p(0, 2), color=BLACK, radius=0.12))
        hollow_dots.add(Dot(numberplane.c2p(2, 2), color=BLACK, radius=0.12))
        hollow_dots.add(Dot(numberplane.c2p(2, 2), color=WHITE, radius=0.06))
    
        self.add(numberplane, steps, filled_dots, hollow_dots)
```
