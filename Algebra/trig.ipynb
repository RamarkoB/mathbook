```python
from manim import *
import numpy as np
```

```python
# scale = 10
scale = 6
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = round(1920 * 4 / (3 * PI))
```

```python
%%manim -v WARNING SineLine

config.frame_width = 3.1 * PI

class SineLine(Scene):
    def construct(self):
        self.camera.background_color = WHITE

        a = MathTex(r"\frac{\pi}{2}")
        b = MathTex(r"\pi")
        c = MathTex(r"\frac{3\pi}{2}")
        d = MathTex(r"2\pi")
        e = MathTex(r"\frac{5\pi}{2}")
        f = MathTex(r"3\pi")


        numberplane = NumberPlane(
            x_range = (0, 3*PI, PI/2),
            y_range = (-2, 2, 1),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": BLACK,
                "stroke_width": 2,
                "stroke_opacity": 0.2
        })
        self.add(numberplane)

        numberplane.add_coordinates({PI/2:a, PI:b,3 * PI/2:c, 2 * PI:d, 5 * PI/2:e}, [-1, 0,1], None)
        numberplane.color = BLACK
        
        numberplane.coordinate_labels[0][0].font_size = 18
        numberplane.coordinate_labels[0][0].next_to((-PI, 0, 0), .74 * DOWN)
        numberplane.coordinate_labels[0][1].next_to((-PI/2, 0, 0), .65 * DOWN)
        numberplane.coordinate_labels[0][1].font_size = 18
        numberplane.coordinate_labels[0][2].next_to((0, 0, 0), .35 * DOWN)
        numberplane.coordinate_labels[0][2].font_size = 18
        numberplane.coordinate_labels[0][3].font_size = 18
        numberplane.coordinate_labels[0][3].next_to((PI/2, 0, 0), .61 * DOWN)
        numberplane.coordinate_labels[0][4].font_size = 18
        numberplane.coordinate_labels[0][4].next_to((PI, 0, 0), .6 * DOWN)

        sine = numberplane.plot(lambda x: np.sin(x), color=BLACK)
        
        numberplane.color = BLACK

        self.add( sine)
```
