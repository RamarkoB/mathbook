```python
from manim import *
import numpy as np
```

```python
scale = 1.5
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING Constant

def func(theta):
    r = 1
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])

class Constant(Scene):
    def construct(self):
        polarplane_pi = PolarPlane(
            azimuth_units="PI radians",
            size = scale * 2,
            radius_step = 1,
            x_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            y_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 1,
                "stroke_opacity": 0.3
            }
        )

        curl = ParametricFunction(func, t_range = np.array([0, TAU]), fill_opacity=0, stroke_width=3).set_color(BLACK)

        self.add(polarplane_pi, curl)
```

```python
%%manim -v WARNING SineLemniscate

def func(theta):
    r = np.sqrt(np.sin(2 * theta))
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])



class SineLemniscate(Scene):
    def construct(self):
        polarplane_pi = PolarPlane(
            azimuth_units="PI radians",
            size = scale * 2,
            radius_step = 1,
            x_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            y_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 1,
                "stroke_opacity": 0.3
            }
        )

        curl1 = ParametricFunction(
            func, 
            t_range = np.array([0+0.000001, (PI/2) - 0.0000001]), 
            # discontinuities = [0,PI/2],
            # dt=0.1,
            fill_opacity=0, 
            stroke_width=3).set_color(BLACK)

        curl2 = ParametricFunction(
            func, 
            t_range = np.array([PI+0.0000001, (3 * PI/2) - 0.0000001]), 
            fill_opacity=0, 
            stroke_width=3).set_color(BLACK)

        self.add(polarplane_pi, curl1,curl2)
```

```python
%%manim -v WARNING CosineLemniscate

def func(theta):
    r = np.sqrt(np.cos(2 * theta))
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])



class CosineLemniscate(Scene):
    def construct(self):
        polarplane_pi = PolarPlane(
            azimuth_units="PI radians",
            size = scale * 2,
            radius_step = 1,
            x_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            y_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 1,
                "stroke_opacity": 0.3
            }
        )

        curl1 = ParametricFunction(
            func, 
            t_range = np.array([- (PI/4), PI/4]), 
            # discontinuities = [0,PI/2],
            # dt=0.1,
            fill_opacity=0, 
            stroke_width=3).set_color(BLACK)

        curl2 = ParametricFunction(
            func, 
            t_range = np.array([3 * PI / 4 + 0.00001, 5 * PI/4 - 0.00001]), 
            # discontinuities = [0,PI/2],
            # dt=0.1,
            fill_opacity=0, 
            stroke_width=3).set_color(BLACK)

        self.add(polarplane_pi, curl1,curl2)
```
