```python
scale = 6
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING Folium

config.frame_width = 20

# from ./algebra
class Folium(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        self.frame_width = scale * 2
        numberplane = NumberPlane(
            x_range = (-scale * 2, scale * 2, step),
            y_range = (-scale * 2, scale * 2, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": 15,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 18,
                "stroke_opacity": 0.6
            }
        )

        folium = ImplicitFunction(
            lambda x, y: x ** 3 + y ** 3 - 18 * x * y,
            color = BLACK,
            x_range = [-scale - 5, scale + 5],
            y_range = [-scale - 5, scale + 5],
        )
        folium.stroke_width = 32

        self.add(numberplane, folium)
```

```python
%%manim -v WARNING VertStretch

config.frame_width = 20

class VertStretch(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-10, 10, step),
            y_range = (-10, 10, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": 15,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 18,
                "stroke_opacity": 0.6
            }
        )

        cubed = FunctionGraph(
            lambda t: (t - 4) * t * (t + 4) * (1/12),
            color = BLACK
        )
        cubed.stroke_width = 32

        strech = FunctionGraph(
            lambda t: 2 * (t - 4) * t * (t + 4) * (1/12),
            x_range=[-6, 6],
            color = BLACK
        )
        strech.stroke_width = 24
        dashed = DashedVMobject(strech, num_dashes=35, equal_lengths =True)

        self.add(numberplane, cubed, dashed)
```

```python
scale = 1
step = 1

config.background_color = WHITE
config.frame_width = scale * 2
config.pixel_width = 1920
config.pixel_height = 1920
```

```python
%%manim -v WARNING SineLemniscate

def func(theta):
    r = np.sqrt(np.sin(2 * theta))
    return np.array([r * np.cos(theta), r * np.sin(theta), 0])


# from ./polar
class SineLemniscate(Scene):
    def construct(self):
        polarplane_pi = PolarPlane(
            azimuth_units="PI radians",
            size = scale * 2,
            radius_step = 1,
            x_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            y_axis_config = {
                "color": BLACK, 
                "stroke_width": 1.5,
                "stroke_opacity": 0.45
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 1,
                "stroke_opacity": 0.3
            }
        )

        curl1 = ParametricFunction(
            func, 
            t_range = np.array([0+0.000001, (PI/2) - 0.0000001]), 
            # discontinuities = [0,PI/2],
            # dt=0.1,
            fill_opacity=0, 
            stroke_width=3).set_color(BLACK)

        curl2 = ParametricFunction(
            func, 
            t_range = np.array([PI+0.0000001, (3 * PI/2) - 0.0000001]), 
            fill_opacity=0, 
            stroke_width=3).set_color(BLACK)

        self.add(polarplane_pi, curl1, curl2)
```

```python
%%manim -v WARNING HoriHyperbola

config.frame_width = scale * 2

class HoriHyperbola(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        numberplane = NumberPlane(
            x_range = (-scale, scale, step),
            y_range = (-scale, scale, step),
            axis_config = {
                "color": BLACK, 
                "stroke_width": scale
            }, 
            background_line_style = {
                "stroke_color": WHITE,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            })

        circle = ImplicitFunction(
            lambda x,y: 1.5 - (x)**2 + y**2,
            color = BLACK,
            x_range = [-4, 4]
        )
        circle.stroke_width = 12

        asymptotes = VGroup()
        
        line1 = FunctionGraph(lambda x: x, color=BLACK, x_range=[-6,6])
        line1.stroke_width = 9
        asymptote1 = DashedVMobject(line1, num_dashes=35, equal_lengths=True)
        
        
        line2 = FunctionGraph(lambda x: -x, color=BLACK, x_range=[-6,6])
        line2.stroke_width = 9
        asymptote2 = DashedVMobject(line2, num_dashes=35, equal_lengths=True)

        asymptotes.add(asymptote1, asymptote2)


        arrows = VGroup()
        arrows.add(Arrow(start=[2.5,np.sqrt(4.75), 0],end=[3.15, np.sqrt(8.4225), 0], color=BLACK))
        arrows.add(Arrow(start=[2.5,-np.sqrt(4.75), 0],end=[3.15, -np.sqrt(8.4225), 0], color=BLACK))
        arrows.add(Arrow(start=[-2.5,np.sqrt(4.75), 0],end=[-3.15, np.sqrt(8.4225), 0], color=BLACK))
        arrows.add(Arrow(start=[-2.5,-np.sqrt(4.75), 0],end=[-3.15, -np.sqrt(8.4225), 0], color=BLACK))


        self.add(circle, asymptotes)
```
